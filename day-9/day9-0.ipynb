```python
student_scores = {
  "Harry": 81,
  "Ron": 78,
  "Hermione": 99, 
  "Draco": 74,
  "Neville": 62,
}

for key in student_scores:
    if student_scores[key]>=91:
        student_scores[key] = "OUTSTANDING"
    elif student_scores[key]>=81:
        student_scores[key] = "EXCEEDS EXPECTATIONS"
    elif student_scores[key]>=71:
        student_scores[key] = "ACCEPTABLE"
    elif student_scores[key]<=70:
        student_scores[key] = "FAIL"

print(student_scores)
```

```python
# NESTING DICTIONARY IN DICTIONARY
travel_log = {
    "France": {
        "city's_visited ":['paris','lille','dijon'],
        "times visited ": 5
        
        },
    "Germany": {
        "city's_visited ":['berlin','Hamburg','Munich'],
        "times visited ": 10
        
        }
}
# dic = {
#     {

#     },
#     {

#     }
# }
for key1 in travel_log:
    print(f"{key1} = ",travel_log[key1],"\n")
    for key2 in travel_log[key1]:
        print(f"{key2} = ",travel_log[key1][key2])
```

```python
# NESTING DICTIONARY IN DICTIONARY
travel_log = [ 
    {
        "Country" : "France",
        "city's_visited ":['paris','lille','dijon'],
        "times visited ": 5
        
        },
     {
         "Country" : "Germany",
        "city's_visited ":['berlin','Hamburg','Munich'],
        "times visited ": 10
        
        }
]

print([f"{key} = {travel_log[i][key]} \n" for i in range(len(travel_log)) for key in travel_log[i] ])
```

```python
import art

asscii = art.logo

def chosing_winner(dic_val):
    winner = ''
    for key in dic_val:
        for key2 in dic_val:
            if dic_val[key]> dic_val[key2]:
                winner = key
    return winner

dic_infos = {}
is_on = True
while is_on:
    print(asscii)
    print("Welcome to the secret auction program.")
    user_name = input("What is your name?\n--> ")
    user_bid = float(input("What's your bid?\n--> "))
    dic_infos[user_name] = user_bid
    another_bidder = input("are there any other bidders? type 'yes' or 'no'\n").lower()
    if another_bidder == 'no' or another_bidder =='n':
        winner_is = chosing_winner(dic_infos)
        print(f"the highest bid is {dic_infos[winner_is]}and the bidder is {winner_is}")
        is_on = False


```
