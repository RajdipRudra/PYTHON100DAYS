a[::-1] EXPLAINATION

```python
word = "DUMBaSSS BiTCH"
#       012345678910-11-12-13
print(word[:])
#word[{starting point}:{ending point}:{}]--> since we left the starting point and the ending point empty. python intrepeter took as the whole string. 
# AND IN THE STEP'S PART WE GAVE -1 SO IT'S GONNA BE COUNTING BACKWARD STEPPING BACK -1 EACH TIME!!
#SO AT FIRST IT WAS [::-1] WHICH IS D THEN [::-2] WHICH IS L AND SO ON AND SO FORTH

print(word[::-1])
```

```python
#we can even skip letters by increasing the steps
print(word[::-3])
```

<h5>YOUR LIFE IN WEEKS<h5>

```python
user_age = int(input("enter ur age : "))

years_left = 90-user_age
months_left = years_left*12
weeks_left = years_left*52
days_left = years_left*365


print(f"You have {days_left} days, {weeks_left} weeks, and {months_left} months left.")
```
