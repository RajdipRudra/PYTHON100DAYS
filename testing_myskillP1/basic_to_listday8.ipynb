```python
rows = 10  # this will control the top half

# total height = top half + middle + bottom half
total_rows = rows * 2 + 1

j = 1  # start stars
direction = 1  # 1 = growing, -1 = shrinking

for i in range(total_rows):
    # print spaces
    print("_" * (rows - j + 1) + "*" * (2 * j - 1))
    
    # change direction at the widest point
    if j == rows + 1:
        direction = -1
    j += direction
```
